## Size of GeoFabric outputs
Estimate the rough size of DEm, Roughness and any other geofabric outputs

```python
nz_area_msq = 268_021_000_000
nz_water_buffer_ratio = 1.1
res_50msq = 50 ** 2
res_10msq = 10 ** 2
res_5msq = 5 ** 2
res_2_5msq = 2.5 ** 2
float_32 = 4
```

```python
def bytes_to_gigabyte(byte_var):
    return byte_var / 1024 / 1024 / 1024
```

# 10m Square for all of NZ

```python
res = 10
res_square = res ** 2
pixels = nz_area_msq / res_square
gb = bytes_to_gigabyte(pixels * float_32)
print(f"NZ land at {res}m square will be {int(pixels)} pixels, or an uncompressed at {round(gb, 2)}GB")
print(f"NZ land plus 10% for water at {res}m square will be {int(nz_water_buffer_ratio * pixels)},"
      f" pixels, or an uncompressed at {round(nz_water_buffer_ratio * gb, 2)}GB")
```
